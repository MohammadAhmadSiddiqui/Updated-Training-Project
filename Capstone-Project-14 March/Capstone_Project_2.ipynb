Task 2 –

Prompt Engineering Experiments (8 Marks)

Design and test three different prompting strategies.

1️⃣ Zero-Shot Prompt
Ask the model to summarize the article without providing examples.

Example task:

Summarize the following article in 5 bullet points.

2️⃣ Few-Shot Prompt
Provide 2 example summaries before asking the model to summarize a new article.

3️⃣ Chain-of-Thought Prompt
Ask the model to reason step-by-step before generating the summary.

Example structure:

Analyze the article step by step:

1. Identify the main topic

2. Extract key ideas

3. Generate a concise summary

Deliverable:

All three prompts


Model outputs


Comparison of responses


```python
from transformers import pipeline

# Load model
generator = pipeline("text-generation", model="gpt2")

# Sample article
article = """
Artificial Intelligence (AI) is transforming industries such as healthcare, education,
and finance. In healthcare, AI helps doctors diagnose diseases using medical images
and predictive analytics. In education, AI systems personalize learning based on
student performance. Financial institutions use AI for fraud detection and automated
trading. Despite these benefits, AI raises concerns related to privacy, bias in
algorithms, and job loss due to automation.
"""

# -------- 1. Zero-Shot Prompt --------
zero_shot_prompt = f"""
Summarize the following article in 5 bullet points:

{article}
"""

zero_output = generator(zero_shot_prompt, max_length=150)[0]['generated_text']


# -------- 2. Few-Shot Prompt --------
few_shot_prompt = f"""
Example 1
Article: AI improves medical diagnosis and helps doctors detect diseases earlier.
Summary:
- AI assists doctors in diagnosis
- Improves disease detection
- Uses medical data analysis

Example 2
Article: Online education platforms use AI to personalize learning for students.
Summary:
- AI analyzes student performance
- Provides personalized learning
- Improves education outcomes

Now summarize the following article in bullet points:

Article:
{article}

Summary:
"""

few_output = generator(few_shot_prompt, max_length=200)[0]['generated_text']


# -------- 3. Chain-of-Thought Prompt --------
cot_prompt = f"""
Analyze the article step by step.

1. Identify the main topic
2. Extract key ideas
3. Generate a concise summary

Article:
{article}

Answer:
"""

cot_output = generator(cot_prompt, max_length=200)[0]['generated_text']
```

```python
print("ZERO SHOT OUTPUT:\n", zero_output)
print("\nFEW SHOT OUTPUT:\n", few_output)
print("\nCHAIN OF THOUGHT OUTPUT:\n", cot_output)
```

Task 3 –
Prompt Optimization (7 Marks)

Improve the prompt so that the AI generates better executive summaries.

Your optimized prompt must produce:

3 key insights


1 actionable takeaway


Professional tone


Test the optimized prompt on two different articles.

Deliverable:

Final optimized prompt


Two example outputs

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")
prompt_template = """
You are an AI Research Assistant.

Read the article and produce a structured executive summary.

Format strictly like this:

Executive Summary

Key Insights:
1.
2.
3.

Actionable Takeaway:

Article:
"""

#Article 1
article1 = """
Artificial Intelligence is transforming healthcare by enabling faster diagnosis,
improving medical imaging analysis, and predicting diseases using large datasets.
Hospitals are increasingly adopting AI-based systems to support doctors in making
more accurate clinical decisions. However, ethical concerns such as data privacy
and algorithmic bias remain major challenges.
"""

prompt1 = prompt_template + article1 + "\nExecutive Summary:\n"

output1 = generator(
    prompt1,
    max_length=200,
    num_return_sequences=1,
    temperature=0.3,
    truncation=True
)


#  Article 2
article2 = """
Online learning platforms are rapidly growing due to the integration of artificial
intelligence. AI helps personalize learning experiences by analyzing student
performance and recommending suitable learning materials. This improves student
engagement and learning outcomes. However, institutions must ensure accessibility
and equal opportunities for all students.
"""

prompt2 = prompt_template + article2 + "\nExecutive Summary:\n"

output2 = generator(
    prompt2,
    max_length=200,
    num_return_sequences=1,
    temperature=0.3,
    truncation=True
)

```

```python
print("OUTPUT 1:")
print(output1[0]["generated_text"])
print("OUTPUT 2:")
print(output2[0]["generated_text"])
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
text = """
Artificial Intelligence is transforming industries such as healthcare,
education, and finance. AI systems analyze large datasets to improve
decision-making and automate complex tasks.
"""

tokens = tokenizer.tokenize(text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)
token_count = len(tokens)

```

```python
print("Original Text:\n", text)
print("\nTokenized Output:\n", tokens)
print("\nToken IDs:\n", token_ids)
print("\nTotal Token Count:", token_count)
```

```python
from transformers import pipeline

from transformers import pipeline

# Use supported pipeline
generator = pipeline("text-generation", model="gpt2")

def ai_research_assistant(article_text):

    prompt = f"""
Summarize the following article.

Return the answer strictly in this format:

Summary:
<short summary>

Key Insights:
1.
2.
3.

Actionable Recommendation:

Article:
{article_text}

Answer:
"""

    result = generator(
        prompt,
        max_length=180,
        temperature=0.3,
        num_return_sequences=1,
        truncation=True
    )

    return result[0]["generated_text"]


# Example article
article = """
Artificial Intelligence is rapidly transforming the healthcare industry.
AI-powered tools help doctors analyze medical images, predict diseases,
and improve diagnosis accuracy. Hospitals are increasingly adopting
machine learning systems to support clinical decision-making.
However, concerns such as patient data privacy and algorithmic bias
must be addressed for responsible AI deployment.
"""

output = ai_research_assistant(article)

print("\n===== AI Assistant Output =====\n")
print(output)

def ai_research_assistant(article_text):

    prompt = f"""
You are an AI Research Assistant.

Read the following article and produce:

1. A short summary
2. Three key insights
3. One actionable recommendation

Format:

Summary:
<short summary>

Key Insights:
1.
2.
3.

Actionable Recommendation:

Article:
{article_text}

Answer:
"""

    result = generator(prompt, max_length=200, temperature=0.3)
    return result[0]["generated_text"]
```
